```python
# https://platform.olimpiada-ai.ro/problems/21

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
```

```python
train = pd.read_csv("/kaggle/input/solar-energy/train.csv")
test = pd.read_csv("/kaggle/input/solar-energy/test.csv")

len(train), len(test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

features = [c for c in train.columns if c not in ['SampleID', 'energy_output']]

X, y = train[features].values, train['energy_output'].values
X_test = test[features].values

scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
X_test = scaler.transform(X_test)

X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
from sklearn.neural_network import MLPRegressor

model = MLPRegressor(max_iter=3100, random_state=42)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error as mae

y_pred = model.predict(X_valid).flatten()

score = mae(y_valid, y_pred, squared=False)

print(f"Score: {score:.5f}")
```

```python
y_pred = model.predict(X_test).flatten()

subm = pd.DataFrame({
    'SampleID': test['SampleID'],
    'energy_output': y_pred
})

subm.head()
```

```python
subm.to_csv("submission.csv", index=False)
```
