```python
# https://www.kaggle.com/competitions/cyprus-ai-camp-2026-tasks

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
import warnings
warnings.filterwarnings('ignore')
```

```python
train_dir = "/kaggle/input/cyprus-ai-camp-2026-tasks/data/train"
test_dir = "/kaggle/input/cyprus-ai-camp-2026-tasks/data/test"
```

```python
candidate_labels = [
    'East Asian Pagoda / Temple Complex.',
    'Colonial Civic Building (Neoclassical).',
    'Ancient Pyramid / Step Pyramid.',
    'Contemporary Deconstructivist Skyscraper.',
    'Mughal Mausoleum / Indo-Islamic Tomb.',
    'Gothic Cathedral.'
]

cand2label = dict([(k, v) for v, k in enumerate(candidate_labels)])
label2cand = dict([(k, v) for k, v in enumerate(candidate_labels)])
```

```python
from transformers import pipeline

model_checkpoint = "geolocal/StreetCLIP"

pipe = pipeline("zero-shot-image-classification", model=model_checkpoint)
```

```python
subm = pd.read_csv("/kaggle/input/cyprus-ai-camp-2026-tasks/data/sample_submission.csv")
subm.head()
```

```python
answers = []
for i, row in tqdm(subm.iterrows(), total=len(subm)):
    answers.append(
        cand2label[pipe(
            f'/kaggle/input/cyprus-ai-camp-2026-tasks/data/{row['test_image_path']}',
            candidate_labels=candidate_labels
        )[0]['label']]
    )

subm['class'] = answers
```

```python
subm.to_csv("submission.csv", index=False)

subm.head()
```
