```python
data = pd.read_csv("/kaggle/input/datasets/abukanabek/museums-romania-9-10/muzee.csv")
subm = pd.read_csv("/kaggle/input/datasets/abukanabek/museums-romania-9-10/sample_output.csv")

data.shape, subm.shape
```

```python
subm.loc[0:3, 'answer'] = [
    len(data),
    (data['județul']=='București').sum(),
    len(data.columns[(data.isna().sum() > 0).values]),
    int(data['anul înființării'].value_counts().index[0])
]
```

```python
subm.iloc[4:4+data['județul'].nunique(), 2] = list(
    map(
        data['județul'].value_counts().to_dict().get, subm.iloc[4:4+42, 0]
    )
)
```

```python
all(subm.iloc[4+data['județul'].nunique():4+data['județul'].nunique()+data.shape[0]]['id'].values == data['_id'].astype('str').values)
```

```python
subm.iloc[4+data['județul'].nunique():4+data['județul'].nunique()+data.shape[0], 2] = (
    (
        (
            data.shape[1] - data.isna().sum(axis=1)
        ) 
        / data.shape[1]
    )
    .map(lambda x: round(100*x, 2))
)
```

```python
subm.iloc[-2:, 2] = [
    round(
        subm
        .iloc[4+data['județul'].nunique():4+data['județul'].nunique()+data.shape[0], 2]
        .mean(), 
        2
    ),
    round(
        (
            subm.iloc[4+data['județul'].nunique():4+data['județul'].nunique()+data.shape[0], 2]
            == 
            (
                subm
                .iloc[4+data['județul'].nunique():4+data['județul'].nunique()+data.shape[0], 2]
                .max()
            )
        ).mean()*100, 2
    )
]
```
