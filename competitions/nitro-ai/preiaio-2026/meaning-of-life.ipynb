```python
# https://judge.nitro-ai.org/competitions/nitro/pre-iaio-2026/3/view

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
from sklearn.decomposition import PCA
```

```python
train = pd.read_csv("/kaggle/input/datasets/abukanabek/preiaio-2026-3/train_data.csv")
subm = pd.read_csv("/kaggle/input/datasets/abukanabek/preiaio-2026-3/sample_output.csv")

train.shape, subm.shape
```

```python
train.head()
```

```python
def choose(X, available, n_subspace_dims=3, epochs=2000, to_keep=20):
    best_choice = None 
    for epoch in tqdm(range(epochs), leave=False):
        sample_indices = np.random.choice(len(available), size=4, replace=False)
        pca = PCA(n_components=n_subspace_dims)

        pca.fit(X[available[sample_indices]])

        reconstruction = pca.inverse_transform(pca.transform(X))

        residuals = np.linalg.norm(X - reconstruction, axis=1)

        new_available = np.argsort(residuals)[:to_keep]
        res_sum = residuals[new_available].sum()
        if best_choice is None or res_sum < best_choice[1]:
            best_choice = (pca, res_sum, new_available)
    print(f'Best score here: {best_choice[1]:.5f}')
    # print(f'New available: {[x.item() for x in sorted(new_available)]}')
    return best_choice
```

```python
X = train[[f'X{i}' for i in range(1, 10+1)]].values
available = np.arange(X.shape[0])
choice = None

for _ in range(10):
    best_choice = choose(X, available)
    if choice is None or choice[1] > best_choice[1]:
        choice = best_choice
        available = best_choice[2]
```

```python
subm.iloc[available + 1, 2] = 1

subm.to_csv("submission.csv", index=False)

subm
```
